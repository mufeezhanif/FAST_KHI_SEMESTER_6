```python

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('wordnet')
```

```python
text = """As the runners were jogging through the hills, they noticed a group of children who were playing near the streams.
The children seemed happier than they had been yesterday, as the weather had been raining then, and the fields had become
too muddy to play. Meanwhile, the runners continued their journey, feeling stronger with each step."""
```

```python
words=text.split(" ")
```

```python
print(words)
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()


for word in words:
    print(word+' -> '+porter_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
snowball_stemmer = SnowballStemmer(language='english')


for word in words:
    print(word+' -> '+snowball_stemmer.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for word in words:
    print(word+' -> '+lemmatizer.lemmatize(word))
```

```python
word_list =['Running', 'Better', 'Caring', 'Wolves',  'Geese', 'Children', 'Went', 'Studies', 'Fancier', 'Mice', 'Flying', 'Bigger','Spoke', 'Women', 'Unhappily']
```

<H2> These words include various linguistic challenges, such as:<H2>


1. Comparative and superlative forms ("Better," "Bigger")
2. Plurals and irregular plurals ("Wolves," "Geese," "Mice")
3. Derivational morphology ("Caring," "Studies")
4. Context-sensitive words ("Spoke," "Flying")
5. Verb conjugations ("Running," "Went")




```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

from nltk.stem.snowball import SnowballStemmer
snowball_stemmer = SnowballStemmer(language='english')

from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()

print("word" +"  ->   " +"Lemmatizer"+ "  ->   "+ "Porter Stemmer"+ "  ->   " +"Snowball Stemmer")
for word in word_list:
    print(word+' ->    '+lemmatizer.lemmatize(word) + " ->    " +porter_stemmer.stem(word)+ "->   " +snowball_stemmer.stem(word))


```
