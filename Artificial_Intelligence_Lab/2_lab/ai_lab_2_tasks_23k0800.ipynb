```python

class SecurityAgent:
    def __init__(self, agent_id, name, status="active"):
        self.agent_id = agent_id
        self.name = name
        self.status = status

    def display_info(self):
        print(f"Agent ID: {self.agent_id}, Name: {self.name}, Status: {self.status}")

class FirewallAgent(SecurityAgent):
    def __init__(self, agent_id, name, rule_set):
        super().__init__(agent_id, name, "active")
        self.rule_set = rule_set

    def monitor_traffic(self):
        print(f"{self.name} (ID: {self.agent_id}) is monitoring network traffic using rule set: {self.rule_set}.")

class MalwareDetectionAgent(SecurityAgent):
    def __init__(self, agent_id, name, virus_db_version):
        super().__init__(agent_id, name, "active")
        self.virus_db_version = virus_db_version

    def scan_files(self):
        print(f"{self.name} (ID: {self.agent_id}) is scanning files for threats using DB version {self.virus_db_version}.")

class AutomationAgent(SecurityAgent):
    def __init__(self, agent_id, name, automation_script):
        super().__init__(agent_id, name, "active")
        self.automation_script = automation_script

    def run_automation(self):
        print(f"{self.name} (ID: {self.agent_id}) is running automation script: {self.automation_script}.")

# Create objects and call methods
firewall = FirewallAgent("FA001", "PrimaryFirewall", "VPC-Rules-2023")
malware_detector = MalwareDetectionAgent("MD002", "EndpointScanner", "v2023.10.26")
automation_bot = AutomationAgent("AA003", "SecurityOptimizer", "optimize_system.py")

firewall.display_info()
firewall.monitor_traffic()

malware_detector.display_info()
malware_detector.scan_files()

automation_bot.display_info()
automation_bot.run_automation()


```

```python
class CybersecurityThreat:
    def __init__(self, threat_id, name, severity="medium"):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity

    def get_info(self):
        return f"Threat ID: {self.threat_id}, Name: {self.name}, Severity: {self.severity}"

class PhishingThreat(CybersecurityThreat):
    def __init__(self, threat_id, name, sender_domain):
        super().__init__(threat_id, name, "high")
        self.sender_domain = sender_domain

    def analyze_email(self):
        print(f"Analyzing email for {self.name} (ID: {self.threat_id}) from domain: {self.sender_domain}.")

class RansomwareThreat(CybersecurityThreat):
    def __init__(self, threat_id, name, encryption_type):
        super().__init__(threat_id, name, "critical")
        self.encryption_type = encryption_type

    def scan_files(self):
        print(f"Scanning file systems for {self.name} (ID: {self.threat_id}) with encryption type: {self.encryption_type}.")

class BotnetThreat(CybersecurityThreat):
    def __init__(self, threat_id, name, C2_server_ip):
        super().__init__(threat_id, name, "high")
        self.C2_server_ip = C2_server_ip

    def detect_traffic(self):
        print(f"Detecting suspicious network traffic for {self.name} (ID: {self.threat_id}) linked to C2 server: {self.C2_server_ip}.")

# Instantiate objects and call methods
phishing = PhishingThreat("PT001", "Spear Phishing Campaign", "evil.com")
ransomware = RansomwareThreat("RT002", "WannaCry Variant", "AES-256")
botnet = BotnetThreat("BT003", "Mirai Botnet", "192.168.1.1")

print(phishing.get_info())
phishing.analyze_email()

print(ransomware.get_info())
ransomware.scan_files()

print(botnet.get_info())
botnet.detect_traffic()

```

```python
class ResponseAgent:
    def __init__(self, agent_id, name):
        self.agent_id = agent_id
        self.name = name

    def execute_response(self):
        raise NotImplementedError("Subclasses must implement this method")

class AlertAgent(ResponseAgent):
    def __init__(self, agent_id, name, notification_channel):
        super().__init__(agent_id, name)
        self.notification_channel = notification_channel

    def execute_response(self):
        print(f"AlertAgent {self.name} (ID: {self.agent_id}) is sending notification via {self.notification_channel}.")

class BlockAgent(ResponseAgent):
    def __init__(self, agent_id, name, firewall_rule_id):
        super().__init__(agent_id, name)
        self.firewall_rule_id = firewall_rule_id

    def execute_response(self):
        print(f"BlockAgent {self.name} (ID: {self.agent_id}) is blocking malicious activity using firewall rule: {self.firewall_rule_id}.")

class RecoverAgent(ResponseAgent):
    def __init__(self, agent_id, name, backup_version):
        super().__init__(agent_id, name)
        self.backup_version = backup_version

    def execute_response(self):
        print(f"RecoverAgent {self.name} (ID: {self.agent_id}) is restoring affected systems from backup version: {self.backup_version}.")

# Demonstrate polymorphism
response_agents = [
    AlertAgent("RA001", "EmailNotifier", "email"),
    BlockAgent("RA002", "IPBlocker", "FW-RULE-001"),
    RecoverAgent("RA003", "SystemRestorer", "20231026-01"),
    AlertAgent("RA004", "SMSNotifier", "SMS")
]

for agent in response_agents:
    agent.execute_response()

```

```python
class Employee:
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id

    def work(self):
        print(f"{self.name} (ID: {self.employee_id}) is performing general work tasks.")

class Manager(Employee):
    def __init__(self, name, employee_id, department):
        super().__init__(name, employee_id)
        self.department = department

    def work(self):
        print(f"Manager {self.name} (ID: {self.employee_id}) is managing department {self.department} and coordinating teams.")

class Developer(Employee):
    def __init__(self, name, employee_id, programming_language):
        super().__init__(name, employee_id)
        self.programming_language = programming_language

    def work(self):
        print(f"Developer {self.name} (ID: {self.employee_id}) is writing code in {self.programming_language}.")

class Designer(Employee):
    def __init__(self, name, employee_id, design_tool):
        super().__init__(name, employee_id)
        self.design_tool = design_tool

    def work(self):
        print(f"Designer {self.name} (ID: {self.employee_id}) is creating designs using {self.design_tool}.")

# Create objects and call work() to show polymorphism
employees = [
    Manager("Alice", "E001", "Engineering"),
    Developer("Bob", "E002", "Python"),
    Designer("Charlie", "E003", "Figma"),
    Employee("David", "E004") # Generic employee
]

for emp in employees:
    emp.work()
```

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.__grade = None # Private attribute for encapsulation

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
            print(f"Grade for {self.name} set to {grade}.")
        else:
            print("Error: Grade must be between 0 and 100.")

    def get_grade(self):
        if self.__grade is not None:
            return self.__grade
        else:
            return "Grade not yet assigned."

    def display_info(self):
        print(f"Student Name: {self.name}, Student ID: {self.student_id}, Grade: {self.get_grade()}")

# Create student objects, update grades, and retrieve them
student1 = Student("Eve", "S101")
student2 = Student("Frank", "S102")

student1.display_info()
student2.display_info()

student1.set_grade(85)
student2.set_grade(92)

student1.display_info()
student2.display_info()

print(f"Eve's grade: {student1.get_grade()}")
print(f"Frank's grade: {student2.get_grade()}")

student1.set_grade(105) # Attempt to set an invalid grade

# Attempting to directly access the private attribute (will work but is discouraged and shows encapsulation purpose)
# print(f"Direct access to Eve's grade (discouraged): {student1._Student__grade}")
```
